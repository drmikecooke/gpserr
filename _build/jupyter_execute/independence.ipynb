# Independence of variables

Intuitively the length of the steps should be at least approximately independent in the probability sense. But there is a niggling worry . . .

```python
import numpy as np
rng = np.random.default_rng()
S=[]
M=1000
for i in range(M):
  N=3
  X=rng.standard_normal(N)
  Y=rng.standard_normal(N)
  S+=[[((X[k+1]-X[k])**2+(Y[k+1]-Y[k])**2)**(1/2) for k in range(N-1)]]
```

```python
import matplotlib.pyplot as plt
X=[s[0] for s in S]
Y=[s[1] for s in S]
plt.plot(X,Y, 'ro')
plt.show()
```

Long steps are correlated with long steps, we fear. Two successive steps share an evaluation point, creating potential dependence!

Let us look at a sequence of three evaluations, $\mathbf r_0,\mathbf r_1,\mathbf r_2$. The apparent total distance is $|\mathbf r_1-\mathbf r_0|+|\mathbf r_2-\mathbf r_1|$.

We can't apply the simple separation achieved with one step. The two parts depend on $\mathbf r_1$.

Let's be simple minded: $\Delta\mathbf r_n=\mathbf r_n-\mathbf r_{n-1}$. Then, $\mathbf r_0=\mathbf r_1-\Delta\mathbf r_1$, $\mathbf r_2=\mathbf r_1+\Delta\mathbf r_2$

The exponential argument in the product of Gaussian distributions will be:

$$\mathbf r_0^2+\mathbf r_1^2+\mathbf r_2^2=3\mathbf r_1^2+\Delta\mathbf r_1^2+\Delta\mathbf r_2^2+2\mathbf r_1.(\Delta\mathbf r_2-\Delta\mathbf r_1)$$

We shift $\mathbf r_1$ by completing the square with a view to its removal by integration:


$$\mathbf r_0^2+\mathbf r_1^2+\mathbf r_2^2=3\left(\mathbf r_1+\frac{\Delta\mathbf r_2-\Delta\mathbf r_1}3\right)^2-\frac{(\Delta\mathbf r_2-\Delta\mathbf r_1)^2}3+\Delta\mathbf r_1^2+\Delta\mathbf r_2^2$$

The Jacobian is 1 for the transformation $d^2\mathbf r_0d^2\mathbf r_1d^2\mathbf r_2=d^2\mathbf r_1d^2\Delta\mathbf r_1d^2\Delta\mathbf r_2$

We integrate out the $\mathbf r_1$ dependence:

$$\int p(\mathbf r_i)d^6\mathbf r_i=\int\exp\left[-\frac{\Delta\mathbf r_1^2+\Delta\mathbf r_2^2+\Delta\mathbf r_1.\Delta\mathbf r_2}{3\sigma^2}\right]\frac13\frac{d^2\Delta\mathbf r_1d^2\Delta\mathbf r_1}{(2\pi\sigma^2)^2}$$

The correlation can be seen from minimizing the exponent. Apart from favouring small displacements, it also tends to anti-correlate successive vectors.
